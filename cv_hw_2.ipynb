```python
# Pre-segmentation: Remove background

import cv2
import numpy as np
from PIL import Image
from pathlib import Path
import matplotlib.pyplot as plt
from segment_anything import sam_model_registry, SamAutomaticMaskGenerator


MASK_INDEX = 1
IMAGE_DIR = "/content/data/train_dataset/custom_dataset/colmap_undistorted/images"
OUTPUT_DIR = "/content/data/train_dataset/custom_dataset/colmap_undistorted/images_masked"
SAM_CHECKPOINT = "/content/SegAnyGAussians/third_party/segment-anything/sam_ckpt/sam_vit_h_4b8939.pth"

Path(OUTPUT_DIR).mkdir(exist_ok=True)

print("Loading SAM model...")
sam = sam_model_registry["vit_h"](checkpoint=SAM_CHECKPOINT)
sam.to(device="cuda")
mask_generator = SamAutomaticMaskGenerator(
    model=sam,
    points_per_side=32,
    pred_iou_thresh=0.86,
    stability_score_thresh=0.92,
    min_mask_region_area=100,
)
print("✓ SAM loaded\n")

image_paths = sorted(Path(IMAGE_DIR).glob("*"))
image_paths = [p for p in image_paths if p.suffix.lower() in ['.png', '.jpg', '.jpeg']]
print(f"Found {len(image_paths)} images\n")


# Masks generation
print("Processing first frame...")
first_image = cv2.imread(str(image_paths[0]))
first_image_rgb = cv2.cvtColor(first_image, cv2.COLOR_BGR2RGB)
masks = mask_generator.generate(first_image_rgb)
print(f"Found {len(masks)} masks\n")

masks_sorted = sorted(masks, key=lambda x: x['area'], reverse=True)[:10]

fig, axes = plt.subplots(2, 5, figsize=(20, 8))
axes = axes.flatten()
for i, mask in enumerate(masks_sorted):
    masked_img = first_image_rgb.copy()
    masked_img[~mask['segmentation']] = 255

    axes[i].imshow(masked_img)
    axes[i].set_title(f"Mask {i}: Area={mask['area']}", fontsize=12, color='red' if i == MASK_INDEX else 'black')
    axes[i].axis('off')

plt.tight_layout()
plt.savefig("/content/mask_options.png", dpi=150, bbox_inches='tight')
plt.show()

print("="*60)
print(f"USING MASK {MASK_INDEX}")
print("="*60)

best_mask = masks_sorted[MASK_INDEX]['segmentation']

fig, axes = plt.subplots(1, 2, figsize=(12, 6))
axes[0].imshow(first_image_rgb)
axes[0].set_title("Original")
axes[0].axis('off')

masked_preview = first_image_rgb.copy()
masked_preview[~best_mask] = 255
axes[1].imshow(masked_preview)
axes[1].set_title(f"Selected Mask {MASK_INDEX}")
axes[1].axis('off')

plt.tight_layout()
plt.savefig("/content/segmentation_preview.png", dpi=150, bbox_inches='tight')
plt.show()
print(f"\n✓ Using mask {MASK_INDEX}\n")

print("Applying segmentation to all frames...")
h, w = first_image_rgb.shape[:2]
reference_bbox = masks_sorted[MASK_INDEX]['bbox']
reference_area = masks_sorted[MASK_INDEX]['area']
reference_center = np.array([
    reference_bbox[0] + reference_bbox[2]/2,
    reference_bbox[1] + reference_bbox[3]/2
])

def find_matching_mask(frame_masks):
    """Finds a mask similar to the reference mask"""
    best_mask = None
    best_score = -1

    for mask in frame_masks:
        bbox = mask['bbox']
        area = mask['area']
        mask_center = np.array([bbox[0] + bbox[2]/2, bbox[1] + bbox[3]/2])

        # Counting similarity
        dist = np.linalg.norm(mask_center - reference_center)
        area_ratio = min(area / reference_area, reference_area / area)

        score = area_ratio / (1 + dist * 0.01)

        if score > best_score:
            best_score = score
            best_mask = mask['segmentation']

    return best_mask if best_mask is not None else np.ones((h, w), dtype=bool)

for i, img_path in enumerate(image_paths):
    image = cv2.imread(str(img_path))
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    frame_masks = mask_generator.generate(image_rgb)
    best_frame_mask = find_matching_mask(frame_masks) if frame_masks else np.ones((h, w), dtype=bool)

    # Apply the mask
    masked_image = image_rgb.copy()
    masked_image[~best_frame_mask] = 255

    # Save
    output_path = Path(OUTPUT_DIR) / img_path.name
    masked_image_bgr = cv2.cvtColor(masked_image, cv2.COLOR_RGB2BGR)
    cv2.imwrite(str(output_path), masked_image_bgr)

    if (i + 1) % 20 == 0:
        print(f"  Processed {i + 1}/{len(image_paths)}")

print(f"\n✓ All frames saved to {OUTPUT_DIR}")
print("\n" + "="*60)
print("Pre-segmentation complete!")
print("Use this command to train:")
print("!python SegAnyGAussians/train_scene.py -s /content/data/train_dataset/custom_dataset/colmap_undistorted --images images_masked --white_background")
print("="*60)
```
