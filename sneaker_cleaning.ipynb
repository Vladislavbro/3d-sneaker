```python
# Filtering by opacity and distance
from plyfile import PlyData, PlyElement
import numpy as np

ply_path = "/content/scene_point_cloud.ply"
plydata = PlyData.read(ply_path)

# Получите данные
xyz = np.stack([plydata['vertex']['x'], plydata['vertex']['y'], plydata['vertex']['z']], axis=1)
opacity = plydata['vertex']['opacity']

# Фильтры
center = xyz.mean(axis=0)
dist = np.linalg.norm(xyz - center, axis=1)
mask = (opacity > -2.0) & (dist < np.percentile(dist, 95))

# Примените фильтр и сохраните
filtered_vertex = plydata['vertex'][mask]
el = PlyElement.describe(filtered_vertex, 'vertex')
PlyData([el]).write('point_cloud_clean.ply')

from google.colab import files
files.download('point_cloud_clean.ply')
```
